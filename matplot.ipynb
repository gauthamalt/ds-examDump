```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3,4,5]
y = [3,10,23,12,20]
```

```python
plt.plot(x,y,marker = 'o',c= "r",markerfacecolor = 'g',linestyle="--")
```

```python
plt.bar(x,y)
```

```python
men = [22,30,35,35,26]
women = [25,32,30,35,29]
```

```python
i = np.arange(len(men))
width = 0.35
```

```python
plt.bar(i,men,width,label="Men")
plt.bar(i+width,women,width,label="Women")
plt.legend()
plt.xticks(i+width/2,('g1','g2','g3','g4','g5'))
```

```python
x = [5,10,20,30,40]
y1 = [1.5,2.5,3.5,4.5,7]
y2 = [5,10.5,2.5,2,5]
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title('graph1')
plt.subplot(2,1,2)
plt.plot(x,y2)
plt.title('graph 2')
plt.show()
```
